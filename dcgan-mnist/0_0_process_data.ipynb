# **Preprocess Data**

This code is used to remove/retain a target class from the MNIST dataset

```python
import os

import torch
import torchvision.datasets as dset
import torchvision.transforms as transforms

import matplotlib.pyplot as plt
```

```python
# Batch size during training
batch_size = 64

# Number of workers for dataloader
workers = 2

# Root directory for dataset
dataroot = "MNIST_full"

# We can use an image folder dataset the way we have it setup.
# Create the dataset
dataset = dset.MNIST(root=dataroot,
                        transform=transforms.Compose([
                            transforms.ToTensor(),
                            transforms.Normalize((0.5,), (0.5,))
                        ]),
                        download=True
                    )

# Create the dataloader
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
target_class = 8

# Use second line if want to remove target class
folder = f"MNIST_{target_class}s/data"
# folder = f"MNIST_without_{target_class}s/data"

os.makedirs(folder, exist_ok=True)

n_targets = 0
for i, (data, label) in enumerate(dataloader):

    images = data[label == target_class]   # Change the boolean operator to `!=` if want to remove target class

    for j, image in enumerate(images):
        image = image.permute(1, 2, 0)
        image = image.numpy()
        image = image.reshape(image.shape[0], image.shape[1])

        file_name = f"{n_targets + 1}.jpg"
        plt.imsave(os.path.join(folder, file_name), image, cmap="gray")
        n_targets += 1

    print(f"[{i + 1}/{len(dataloader)}] # Collected: {n_targets}")
```
